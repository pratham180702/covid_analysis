# Analysis based on population

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

## Data Reading 

```python
df = pd.read_csv('covid_worldwide.csv')
df
```

```python
df.iloc[90,:]
```

```python
# addind china's missing and wrong values
df.at[90, 'Population'] = '1,410,000,000'
df.at[90,'Total Test'] = '2,100,000,000'
df.at[90,'Total Cases'] = '2023378'
df.at[90,'Total Deaths'] = '87,468'
df.at[90,'Total Recovered'] = '1,935,910'
```

```python
#removig countries with 0 population

df = df.dropna(subset=['Population'])
print(df)
```

```python
# dropping all the rows with an empty field

df = df.dropna()
```

```python
df
```

```python
# new location of china

df.iloc[86,:]
```

### removing comas and converting Datatype from object to int 

```python
df['Total Cases'] = df['Total Cases'].replace(',', '', regex=True)
```

```python
df['Total Cases'] = df['Total Cases'].astype(int)
```

```python
df['Total Deaths'] = df['Total Deaths'].replace(',', '', regex=True)
df['Total Deaths'] = df['Total Deaths'].astype(int)

df['Total Recovered'] = df['Total Recovered'].replace(',', '', regex=True)
df['Total Recovered'] = df['Total Recovered'].astype(int)

df['Active Cases'] = df['Active Cases'].replace(',', '', regex=True)
df['Active Cases'] = df['Active Cases'].astype(int)

df['Total Test'] = df['Total Test'].replace(',', '', regex=True)
df['Total Test'] = df['Total Test'].astype(int)

df['Population'] = df['Population'].replace(',', '', regex=True)
df['Population'] = df['Population'].astype(int)
```

```python
df
```

```python
# sorting the values based on population in descending order

df_sorted_2 = df.sort_values('Population', ascending = False)
```

```python
df_sorted_2
```

```python
# calculating percent_affected, percent_recovered, percent_tested

df_sorted_2['percent_affected'] = df_sorted_2['Total Cases'] / df_sorted_2['Population'] * 100

df_sorted_2['percent_recovered'] = df_sorted_2['Total Recovered']/df_sorted_2['Total Cases'] * 100

# df_sorted_2['percent_tested'] = df_sorted_2['Total Test']/df_sorted_2['Population']*100
# df_sorted_2['percent_tested'] = (df_sorted_2['Total Test'] / df_sorted_2['Population'] * 100).clip(upper=100)

df_sorted_2
```

# Data for tests conducted per 1000

```python
tests_df = pd.read_csv('full-list-cumulative-total-tests-per-thousand.csv')
```

```python
tests_df
```

```python
# data is given on the daily basis, so we have to find the mean tests/1000 for each country
# group_by returns a Series so to_frame() converts the series into dataframe
# reset_index() will reset the index of the dataframe

grouped_tests_df = tests_df.groupby("Entity")["total_tests_per_thousand"].mean().to_frame().reset_index()
```

```python
# round(2) will round off the value to 2 decimal places(.2f)

grouped_tests_df["total_tests_per_thousand"] = grouped_tests_df["total_tests_per_thousand"].round(2)
grouped_tests_df
```

```python
# renaming

grouped_tests_df = grouped_tests_df.reset_index().rename(columns={'Entity': 'Country'})
```

```python
# changing the names of the countries to make it compatible with df_sorted to
# FROM United States to USA
# FROM United Kingdom to UK
# FROM South Korea to S. Korea

grouped_tests_df.at[179,'Country']='USA'

grouped_tests_df.at[178,'Country']='UK'

grouped_tests_df.at[158,'Country']='S. Korea'
```

```python
# merging

merge_df = pd.merge(df_sorted_2, grouped_tests_df , on='Country')
merge_df.head()
```

```python
# selecting top 10 countries with highest population

merge_df = merge_df[:10]
merge_df
```

```python
# calculating %

merge_df['percent_tested_per_1000'] = merge_df['total_tests_per_thousand'] * 0.1
```

```python
# selecting required variables

df_final = merge_df[['Serial Number','Country','percent_affected','percent_tested_per_1000','percent_recovered','Population']]
```

```python
# sorting values based on percentage affected

df_final = df_final.sort_values('percent_affected')
df_final
```

### Adding fully vaccinated percent

```python
df_vacc = pd.read_csv('current-global.csv')
```

```python
# changing the name of country from US to USA to make it compatible while merging with df_sorted_2

df_vacc.at[212,'name']='USA'
```

```python
# finding the index of china inorder to change the name

index = df_vacc.index[df_vacc['name'] == 'Mainland China'][0]
index
```

```python
# changing the name from Mainland China to China


df_vacc.at[48,'name']='China'
```

```python
# df_vacc is a Series so converting it to dataframe

df_vacc1 = df_vacc.reset_index().rename(columns={'name': 'Country'})
```

```python
df_vacc2 = df_vacc1[['Country','completedVaccination']]
df_vacc2
```

```python
# merging df_vacc2 with df_sorted_2

vacc_merged_df = pd.merge(df_final, df_vacc2 , on='Country')
vacc_merged_df
```

```python
# calculating fully vaccinated percent %

vacc_merged_df['percent_fully_vacc'] = vacc_merged_df['completedVaccination']/vacc_merged_df['Population'] * 100
vacc_merged_df
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# create a plot with custom parameters
fig, ax = plt.subplots(figsize=(8, 6), dpi=100)

# adding the first plot to the graph
vacc_merged_df.plot('percent_affected', 'percent_recovered',  c='blue', marker='o', ax=ax)

# adding the second plot to the graph
vacc_merged_df.plot('percent_affected', 'percent_tested_per_1000', c='red', marker='s', ax=ax)

#adding the third plot to graph
vacc_merged_df.plot('percent_affected', 'percent_fully_vacc', c='green', marker='p', ax=ax)

# adding text labels and dotted line for each data point
for i, row in vacc_merged_df.iterrows():
    if row['Country'] in ['USA','China']:
        ax.text(row['percent_affected'] +0.125, row['percent_recovered']+6 , f"{row['percent_recovered']:.2f}", fontsize=10)
        ax.text(row['percent_affected'] + 0.25, row['percent_tested_per_1000'], row['Country'], fontsize=10)
        ax.text(row['percent_affected'] + 0.25, row['percent_tested_per_1000']+10, row['Population'], fontsize=10)
        ax.text(row['percent_affected'] + 0.25, row['percent_fully_vacc']-10, f"{row['percent_fully_vacc']:.2f}", fontsize=10)
        ax.vlines(row['percent_affected'], 0, row['percent_tested_per_1000']+40, linestyle='dotted', alpha=0.5,color='red')

# setting plot labels and title
ax.set_xlabel('Percent Affected')
ax.set_ylabel('Percent Recovered with Tested/1000')
ax.set_title('COVID-19 Recovery and Testing Rates by Country Based on population')

# moving the legend to the top left corner
ax.legend(loc='upper right')

plt.show()
```
