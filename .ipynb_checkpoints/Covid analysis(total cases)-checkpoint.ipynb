# Data of Total conducted tested per 1000

```python
tests_df = pd.read_csv('full-list-cumulative-total-tests-per-thousand.csv')
```

```python
tests_df
```

```python
# data is given on the daily basis, so we have to find the mean tests/1000 for each country
# group_by returns a Series so to_frame() converts the series into dataframe
# reset_index() will reset the index of the datafram

grouped_tests_df = tests_df.groupby("Entity")["total_tests_per_thousand"].mean().to_frame().reset_index()
```

```python
# round(2) will round off the value to 2 decimal places(.2f)

grouped_tests_df["total_tests_per_thousand"] = grouped_tests_df["total_tests_per_thousand"].round(2)
grouped_tests_df
```

```python
# renaming

grouped_tests_df = grouped_tests_df.reset_index().rename(columns={'Entity': 'Country'})
```

```python
index = grouped_tests_df.index[grouped_tests_df['Country'] == 'United States'][0]
index
```

```python
index = grouped_tests_df.index[grouped_tests_df['Country'] == 'United Kingdom'][0]
index
```

```python
index = grouped_tests_df.index[grouped_tests_df['Country'] == 'South Korea'][0]
index
```

```python
# changing the names of the countries to make it compatible with df_sorted to
# FROM United States to USA
# FROM United Kingdom to UK
# FROM South Korea to S. Korea

grouped_tests_df.at[179,'Country']='USA'

grouped_tests_df.at[178,'Country']='UK'

grouped_tests_df.at[158,'Country']='S. Korea'
```

```python
# merging

merge_df = pd.merge(df_final, grouped_tests_df , on='Country')
merge_df
```

```python
# calculating %

merge_df['percent_tested_per_1000'] = merge_df['total_tests_per_thousand'] * 0.1
```

```python
merge_df
```

```python
# picking required variables

df_final = merge_df[['Serial Number','Country','percent_affected','percent_tested_per_1000','percent_recovered','Population']]
```

```python
df_vacc2 = df_vacc1[['Country','completedVaccination']]
df_vacc2
```

```python
vacc_merged_df = pd.merge(df_final, df_vacc2 , on='Country')
vacc_merged_df
```

```python
# vacc_merged_df['fully_vaccinated'] = vacc_merged_df['completedVaccination']/10000000
```

```python
vacc_merged_df
```

```python
vacc_merged_df['percent_fully_vacc'] = vacc_merged_df['completedVaccination']/vacc_merged_df['Population'] * 100
vacc_merged_df
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# create a plot with custom parameters
fig, ax = plt.subplots(figsize=(8, 6), dpi=100)

# add the first scatter plot to the graph
vacc_merged_df.plot('percent_affected', 'percent_recovered',  c='blue', marker='o', ax=ax)

# add the second scatter plot to the graph
vacc_merged_df.plot('percent_affected', 'percent_tested_per_1000', c='red', marker='s', ax=ax)

#adding third plot
vacc_merged_df.plot('percent_affected', 'percent_fully_vacc', c='green', marker='p', ax=ax)

# add text labels and dotted line for each data point
for i, row in vacc_merged_df.iterrows():
    if row['Country'] in ['Brazil', 'Italy', 'France']:
        ax.text(row['percent_affected'] , row['percent_recovered'] - 10, f"{row['percent_recovered']:.2f}", fontsize=10)
        ax.text(row['percent_affected'] + 0.5, row['percent_tested_per_1000'], row['Country'], fontsize=10)
        ax.vlines(row['percent_affected'], 0, row['percent_tested_per_1000'], linestyle='dotted', alpha=0.5,color='red')
    else:
        ax.text(row['percent_affected'] - 2.5, row['percent_recovered'] + 5, f"{row['percent_recovered']:.2f}", fontsize=10)
        ax.text(row['percent_affected'] + 1, row['percent_tested_per_1000'], row['Country'], fontsize=10)
        ax.vlines(row['percent_affected'], 0, row['percent_tested_per_1000'], linestyle='dotted', alpha=0.5,color='red')

# set plot labels and title
ax.set_xlabel('Percent Affected')
ax.set_ylabel('Percent Recovered with Tested/1000')
ax.set_title('COVID-19 Recovery and Testing Rates by Country(based on total cases)')

# display the plot
plt.show()
```

```python
# logic implementation

new_vacc_df = vacc_merged_df[vacc_merged_df['percent_recovered']>98]


new_vacc_df = new_vacc_df[vacc_merged_df['percent_tested_per_1000']>=100]
```
