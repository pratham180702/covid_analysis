# Logic Implementation
### if(% of population got +ve by covid is low as possible)
### and 
### % of recovered is as high as possible
### and
### Tests conducted on population is more than 100%(each person is tested atleast once)

## then that country has given tough fight to covid

```python
# calculating the percent affected by covid of the total population

df['percent_affected'] = df['Total Cases'] / df['Population'] * 100
```

```python
df
```

```python
# calculating the percent recovered out of these affected people

df['percent_recovered'] = df['Total Recovered']/df['Total Cases'] * 100
df.head(5)
```

```python
# calculating the percent of people tested for covid out of total population

df['percent_tested'] = df['Total Test']/df['Population']*100
df
```

```python
# Sort the DataFrame by the percent_affected column in ascending order

df_sorted = df.sort_values('percent_affected')
```

```python
df_sorted
```

```python
plt.scatter(df['percent_affected'],df['percent_recovered'])
```

```python
# setting a new index

df_sorted = df_sorted.set_index(np.arange(50))
```

```python
df_sorted
```

```python
df_sorted['Serial Number'] = df_sorted['Serial Number'].astype(int)
```

```python
df_sorted.head(5)
```

```python
# out of all the countries selecting those whose serial number is < 10 means they are in top 10 for total number of cases

df1 = df_sorted[df_sorted['Serial Number'] <= 10]
```

```python
df1
```

```python
# final cleaning of dataset

df_final = df1[['Serial Number','Country','percent_affected','percent_recovered','percent_tested','Population']]
```

```python
vacc_merged_df = pd.merge(df_final, df_vacc2 , on='Country')
vacc_merged_df
```

```python
vacc_merged_df
```

```python
vacc_merged_df['percent_fully_vacc'] = vacc_merged_df['completedVaccination']/vacc_merged_df['Population'] * 100
vacc_merged_df
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# create a plot with custom parameters
fig, ax = plt.subplots(figsize=(8, 6), dpi=100)

# add the first scatter plot to the graph
vacc_merged_df.plot('percent_affected', 'percent_recovered',  c='blue', marker='o', ax=ax)

# add the second scatter plot to the graph
vacc_merged_df.plot('percent_affected', 'percent_tested', c='red', marker='s', ax=ax)

#adding third plot
vacc_merged_df.plot('percent_affected', 'percent_fully_vacc', c='green', marker='p', ax=ax)

# add text labels and dotted line for each data point
for i, row in vacc_merged_df.iterrows():
    if row['Country'] in ['Brazil', 'Italy', 'France']:
        ax.text(row['percent_affected'] , row['percent_recovered'] - 30, f"{row['percent_recovered']:.2f}", fontsize=10)
        ax.text(row['percent_affected'] + 0.5, row['percent_tested'], row['Country'], fontsize=10)
        ax.vlines(row['percent_affected'], 0, row['percent_tested'], linestyle='dotted', alpha=0.5,color='red')
    else:
        ax.text(row['percent_affected'] - 1, row['percent_recovered'] + 20, f"{row['percent_recovered']:.2f}", fontsize=10)
        ax.text(row['percent_affected'] + 1, row['percent_tested'], row['Country'], fontsize=10)
        ax.vlines(row['percent_affected'], 0, row['percent_tested'], linestyle='dotted', alpha=0.5,color='red')

# set plot labels and title
ax.set_xlabel('Percent Affected')
ax.set_ylabel('Percent Recovered with Tested')
ax.set_title('COVID-19 Recovery and Testing Rates by Country(based on total cases)')

# display the plot
plt.show()
```
