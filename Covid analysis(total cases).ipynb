```python
df['percent_affected'] = df['Total Cases'] / df['Population'] * 100
```

```python
df
```

```python
df['percent_recovered'] = df['Total Recovered']/df['Total Cases'] * 100
df.head(5)
```

```python
df['percent_tested'] = df['Total Test']/df['Population']*100
df
```

```python
# Sort the DataFrame by the percent_affected column in ascending order
df_sorted = df.sort_values('percent_affected')

# Print the sorted DataFrame
print(df_sorted)
```

```python
df1 = df_sorted[df_sorted['Serial Number'] <= 10]
```

```python
df_final = df1[['Serial Number','Country','percent_affected','percent_recovered','percent_tested']]
```

```python
df_final
```

```python
# df_final.plot('percent_affected','percent_recovered')
fig, ax = plt.subplots(figsize=(8, 6), dpi=100)
df_final.plot('percent_affected', 'percent_recovered', c='blue', marker='o', ax=ax)

for i, row in df_final.iterrows():
    ax.text(row['percent_affected'] + 1, row['percent_recovered'], row['Country'], fontsize=10)
# set plot labels and title
ax.set_xlabel('Percent Affected')
ax.set_ylabel('Percent Recovered')
ax.set_title('COVID-19 Recovery Rates')

# display the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# create a plot with custom parameters
fig, ax = plt.subplots(figsize=(8, 6), dpi=100)

# add the first scatter plot to the graph
df_final.plot('percent_affected', 'percent_recovered',  c='blue', marker='o', ax=ax)

# add the second scatter plot to the graph
df_final.plot('percent_affected', 'percent_tested', c='red', marker='s', ax=ax)

# add text labels and dotted line for each data point
for i, row in df_final.iterrows():
    if row['Country'] in ['Brazil', 'Italy', 'France']:
        ax.text(row['percent_affected'] + 1, row['percent_recovered'] - 30, f"{row['percent_recovered']:.2f}", fontsize=10)
        ax.text(row['percent_affected'] + 1, row['percent_tested'], row['Country'], fontsize=10)
        ax.vlines(row['percent_affected'], 0, row['percent_tested'], linestyle='dotted', alpha=0.5,color='red')
    else:
        ax.text(row['percent_affected'] - 3, row['percent_recovered'] + 20, f"{row['percent_recovered']:.2f}", fontsize=10)
        ax.text(row['percent_affected'] + 1, row['percent_tested'], row['Country'], fontsize=10)
        ax.vlines(row['percent_affected'], 0, row['percent_tested'], linestyle='dotted', alpha=0.5,color='red')

# set plot labels and title
ax.set_xlabel('Percent Affected')
ax.set_ylabel('Percent Recovered with Tested')
ax.set_title('COVID-19 Recovery and Testing Rates by Country')

# move the legend to the top left corner
ax.legend(loc='upper left')

# display the plot
plt.show()
```
